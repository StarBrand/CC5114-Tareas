## A Circle

```python
from neural_network.utils import Circle

RADIUS = 25

def center():
    return uniform(-10, 10), uniform(-10, 10)

circle = Circle(RADIUS, center(), (MIN_X, MAX_X), (MIN_Y, MAX_Y))
```

```python
x, y = circle.graph()
fig, ax = plt.subplots(figsize=(5, 5))

circle.add_results(ax)

ax.grid()
ax.legend()
```

```python
training_set = circle.training_set(10000)
test_set = circle.training_set(1000)

x1 = []
x2 = []
y = []
x1_sample_in = []
x2_sample_in = []
x1_sample_out = []
x2_sample_out = []

for i, j, k in training_set:
    x1.append(i)
    x2.append(j)
    if k:
        y.append(1.0)
        x1_sample_out.append(i)
        x2_sample_out.append(j)
    else:
        y.append(0.0)
        x1_sample_in.append(i)
        x2_sample_in.append(j)

x_input = np.array([x1, x2])
y_output = np.array([y])
    
trained_parameters = model(x_input, y_output, n_x, n_h, n_y, num_of_iters, learning_rate)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.plot(x1_sample_out, x2_sample_out, "*")
ax.plot(x1_sample_in, x2_sample_in, ".")

ax.grid()
```

```python
from provided_code import result

x1 = []
x2 = []

for i, j, _ in test_set:
    x1.append(i)
    x2.append(j)

fig, ax = plt.subplots(figsize=(5, 5))

x1, y1, x2, y2 = result(predict, trained_parameters, x1, x2, ax)

circle.add_results(ax)

ax.legend()
ax.grid()
```
